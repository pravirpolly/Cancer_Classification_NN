```python
X = df.drop('benign_0__mal_1', axis=1).values
```

```python
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))

model.add(Dense(15, activation='relu'))

# Binary Classification Problem last activation should be sigmoid
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs = 600, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
# graph represents overfitting
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))

model.add(Dense(15, activation='relu'))

# Binary Classification Problem last activation should be sigmoid
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# This callback will stop the training when there is no improvement in the loss for three consecutive epochs.
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose=1, patience=25)
# patience = 25 means that the 25 epochs more will run after the min has been established
```

```python
model.fit(x=X_train, y=y_train, epochs = 600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))
# model.add(Dropout(0.5)) - 0.5 implies 50% of neurons will be dropped out during each run in the layer of 30

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))

# Binary Classification Problem last activation should be sigmoid
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
# This callback will stop the training when there is no improvement in the loss for three consecutive epochs.
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose=1, patience=25)
# patience = 25 means that the 25 epochs more will run after the min has been established
```

```python
model.fit(x=X_train, y=y_train, epochs = 600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
# the model is showing even better behaviour
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
# accuracy is good measure for this model because the cancerous vs benign was a blanced or close to balanced
```

```python
print(confusion_matrix(y_test, predictions))
```
